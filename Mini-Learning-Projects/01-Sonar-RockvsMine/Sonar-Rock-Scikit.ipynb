## Importing Dependencies

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas Dataframe
sonar_data = pd.read_csv('./sonar data.csv', header=None)
```

```python
# How the data looks
sonar_data.head()
```

```python
# number of rows and columns
sonar_data.shape
```

```python
sonar_data.describe() #describe --> statistical measures of the data
```

```python
sonar_data[60].value_counts()
```

```python
df = sonar_data.groupby(60).mean()
df
```

## Plot of Mean Values of each Features

```python
import matplotlib.pyplot as plt

df_transposed = df.T
df_transposed.columns = ['Mines', 'Rock']

fig, ax1 = plt.subplots(figsize=(18, 8))

df_transposed.plot(kind='bar', ax=ax1, width=0.8, position=0.5)

ax2 = ax1.twinx()
df_transposed.plot(kind='line', ax=ax2, marker='o', color=['blue', 'orange'], linewidth=2, linestyle='--')

ax1.set_title('Visualization of Features in Mines and Rock', fontsize=16)
ax1.set_xlabel('Features', fontsize=14)
ax1.set_ylabel('Mean Values', fontsize=14)

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
# separating data and Labels
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]


plt.matshow(X.corr())
plt.show()
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

```python
model = LogisticRegression()
#training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

```python
from sklearn import svm
#training the Support vector classifier model with training data
clf = svm.SVC()
clf.fit(X_train, Y_train)
```

```python
#accuracy on training data using LR
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) 
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on training data using SVC
X_train_prediction = clf.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) 
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data on LR
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test) 
print('Accuracy on test data : ', test_data_accuracy)
```

```python
#accuracy on test data on SVC
X_test_prediction = clf.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test) 
print('Accuracy on test data : ', test_data_accuracy)
```
